```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLOv9e model
model2 = YOLO('yolov9e.pt')

# Train the model on the dataset for 30 epochs
# Note: place path to custom YAML file (created in YOLOv8 notebook) in the 'data' argument below
results2 = model2.train(data='/home/abdulrahman.almarzoo/Desktop/AI702Proj/Drowsiness-/-Fatigue_Detection-4/data.yaml',
    epochs=30, 
    batch=-1, 
    optimizer='AdamW', 
    amp=False, lr0=0.001, 
    momentum=0.9, 
    weight_decay=0.001, 
    label_smoothing=0.1, 
    dropout=0.35
)
```
