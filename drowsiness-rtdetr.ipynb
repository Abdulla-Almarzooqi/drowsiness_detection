```python
from ultralytics import RTDETR

# Load a COCO-pretrained RT-DETR-x model
model = RTDETR('rtdetr-x.pt')

# Train the model on the dataset for 30 epochs
# Note: place path to custom YAML file (created in YOLOv8 notebook) in the 'data' argument below
results = model.train(data='/home/abdulrahman.almarzoo/Desktop/AI702Proj/Drowsiness-/-Fatigue_Detection-4/data.yaml', 
    epochs=30, 
    batch=16, 
    imgsz=640, 
    lr0=0.001, 
    lrf=0.1, 
    momentum=0.9, 
    weight_decay=0.001, 
    label_smoothing=0.1, 
    nbs=16, 
    dropout=0.35
)
```
