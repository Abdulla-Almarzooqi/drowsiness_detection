### Demo for Object Detection Models Used

First load our pretrained models from [here](https://mbzuaiac-my.sharepoint.com/personal/abdulrahman_almarzooqi_mbzuai_ac_ae/_layouts/15/onedrive.aspx?id=%2Fpersonal%2Fabdulrahman%5Falmarzooqi%5Fmbzuai%5Fac%5Fae%2FDocuments%2FBest%20Model%20Weights&ct=1714335263260&or=OWA%2DNT%2DMail&cid=22ae46c4%2D3d0a%2D2e04%2Df78a%2D5b39b386d07d&ga=1&LOF=1). Then place them under `models` folder. Demo images are in the `images` folder. There are four images, their labels are {img1: alert, img2:drowsy, img3: alert, img4: drowsy}.

```python
from ultralytics import YOLO, RTDETR
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Helper method to display predictions
# This method assumes you pass [img1, img2, img3, img4] to the predict() function.
# The argument i represents the number of the folder "predict" under "runs/detect". For example,
# if results are saved to "runs/detect/predict/" then i = 1, if saved to "runs/detect/predict2/"
# then i = 2, and so on. The folder "runs/detect/predict#/" is automatically created upon
# execution of the code.

def display_pred(i):
    images = []

    if i==1:
        for j in range(4):
            path = f"./runs/detect/predict/img{j+1}.jpg"
            images.append(Image.open(path))
            j += 1
    else:
        for j in range(4):
            path = f"./runs/detect/predict{i}/img{j+1}.jpg"
            images.append(Image.open(path))
            j += 1

    fig = plt.figure(figsize=(10, 10))
  
    for i in range(4): 
        fig.add_subplot(2, 2, i+1) 
        plt.imshow(images[i]) 
        plt.axis('off') 
        plt.title(f"img{i+1}")  
    
    plt.tight_layout()
```

```python
# Initialize test images

img1 = "./images/img1.jpg" # alert
img2 = "./images/img2.jpg" # drowsy
img3 = "./images/img3.jpg" # alert
img4 = "./images/img4.jpg" # drowsy
```

#### YOLOv8

```python
# Load our pretrained YOLOv8 model
yolov8 = YOLO("./models/yolov8.pt")
```

```python
# Get predictions
yolov8.predict(source=[img1, img2, img3, img4], verbose=True, save=True)
```

```python
# Display predictions
display_pred(1)
```

#### YOLOv9

```python
# Load our pretrained YOLOv9 model
yolov9 = YOLO("./models/yolov9.pt")
```

```python
# Get predictions
yolov9.predict(source=[img1, img2, img3, img4], verbose=True, save=True)
```

```python
# Display predictions
display_pred(2)
```

```python
# Load our pretrained RT-DETR model
rtdetr = RTDETR("./models/rtdetr.pt")
```

```python
# Get predictions
rtdetr.predict(source=[img1, img2, img3, img4], verbose=True, save=True)
```

```python
# Display predictions
display_pred(3)
```
